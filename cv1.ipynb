```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
```

```python
#drawing
#line: cv2.line(pic variable, start, end, color, thickness)
#rect: cv2.rectangle(pic variable, upper left, lower right, color, thickness)
#circle: cv2.circle(pic variable, center, r, color, thickness)
#thickness -> -1 = filled shape

#canvas
m = np.full((320, 320, 3), (0, 0, 0), np.uint8)
#line
cv2.line(m, (50, 20), (110, 20), (0, 0, 255), 3)
cv2.line(m, (210, 20), (270, 20), (0, 0, 255), 3)
cv2.line(m, (50, 20), (5, 60), (0, 0, 255), 3)
cv2.line(m, (315, 60), (270, 20), (0, 0, 255), 3)
cv2.line(m, (110, 20), (160,60), (0, 0, 255), 3)
cv2.line(m, (210, 20), (160,60), (0, 0, 255), 3)
cv2.line(m, (10, 140), (40, 220), (0, 0, 255), 3)
cv2.line(m, (310,140), (280, 220), (0, 0, 255), 3)
cv2.line(m, (5, 60), (10, 140), (0, 0, 255), 3)
cv2.line(m, (315, 60), (310, 140), (0, 0, 255), 3)
cv2.line(m, (160, 300), (40, 220), (0, 0, 255), 3)
cv2.line(m, (160, 300), (280, 220), (0, 0, 255), 3)
#rect
cv2.rectangle(m, (75, 60), (245, 234), (255, 255, 255), 3)
#circle
cv2.circle(m, (160, 147), 80, (255, 0, 0), 3)
cv2_imshow(m)
```

```python
#儲存影片
#建立儲存控制變數
#控制變數=cv2.VideoWriter("影片路徑", 格式, 每秒影格數, 畫面大小)
p1=cv2.VideoWriter("1.mp4", cv2.VideoWriter_fourcc(*'MP4V'), 30, (300,150))
#cv2.VideoWriter_fourcc(*'編碼格式')
for x in range(1,400,10):
    m1=np.full((150,300,3), (0,0,0), np.uint8)
    m1=Image.fromarray(m1)
    ImageDraw.Draw(m1).text((x,30), "文字", (0,0,255), ImageFont.truetype("kaiu.ttf", 50))
    m1=np.array(m1)
    p1.write(m1) #.write -> 影片寫入格式
p1.release() #釋放控制變數
```

```python
import codecs
import base64

#將影片轉檔為base64格式
with codecs.open("2.mp4","rb") as f:
  f=base64.b64encode(f.read()).decode()
#顯示影片撥放器
dp.HTML(
  f'<video src="data:video/mp4;base64,{f}" controls />'
)
```

```python
#影片讀取
p1=cv2.VideoCapture("1.mp4") # 讀取括號內指定路徑檔案
while p1.isOpened()==True: # 判斷來源是否讀取成功 有被打開
    ret, m1=p1.read() # 是否讀到畫面(布林), 讀到畫面變數 = 控制變數.read
    if ret==True:
        #(1, 3, 4, 5, 7) <-> (當前影格, 影片寬度, 影片高度, 每秒影格數, 影片總影格數)
        print("當前的畫面：",p1.get(1)) # 取得來源資訊
        #.set設定來源資訊
        # dp.clear_output(True)
        cv2.circle(m1, (20,20), 30, (0,255,0), -1)
        cv2_imshow(m1)
        # time.sleep(0.3)
    else:
        break
```
